```python
import matplotlib.pyplot as plt
```

```python
# Figure 객체 생성
fiqure = plt.figure()

axes = fiqure.add_subplot(223) # 행, 열, 번호
x = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
y = [1200, 800, 500, 400, 700, 800]
axes.plot(x, y, linestyle='--', marker='^', color='red')
plt.show()
```
